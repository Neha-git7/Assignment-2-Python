**Question 3**: An e-commerce platform has customer IDs from two campaigns:

campaign1 = {101, 102, 103, 104, 105}

campaign2 = {104, 105, 106, 107}
1. Find unique customers across both campaigns.
2. Find customers common in both.
3. Find customers exclusive to campaign1.
4. Explain how sets help in removing duplicates in real DS projects.


```python
# Customer IDs from campaigns
campaign1 = {101, 102, 103, 104, 105}
campaign2 = {104, 105, 106, 107}

# 1. Unique customers across both campaigns
unique_customers = campaign1.union(campaign2)

print("Unique Customers:")
print(unique_customers)

# 2. Customers common in both campaigns
common_customers = campaign1.intersection(campaign2)

print("\nCommon Customers:")
print(common_customers)

# 3. Customers exclusive to campaign1
exclusive_customers = campaign1.difference(campaign2)

print("\nCustomers Exclusive to campaign1:")
print(exclusive_customers)
```

Explain how sets help in removing duplicates in real DS projects.

-> Sets automatically remove duplicate values, making them very useful in Data Science projects.

- They help clean duplicate records from datasets.
- Set operations like union and intersection make data comparison easy.
- Sets improve performance when handling large amounts of data.
- They are commonly used in customer analysis, recommendation systems, and data preprocessing.

**Question 5**: Given:

sales = [12000, 15000, 11000, 18000, 17000]

Using a loop:
1. Calculate total revenue.
2. Count how many months had sales above 15000.
3. Find maximum sales without using built-in max().


```python
# Sales data
sales = [12000, 15000, 11000, 18000, 17000]

# 1. Calculate total revenue
total_revenue = 0

for amount in sales:
    total_revenue += amount

print("Total Revenue:", total_revenue)

# 2. Count months with sales above 15000
count = 0

for amount in sales:
    if amount > 15000:
        count += 1

print("Months with Sales Above 15000:", count)

# 3. Find maximum sales without using max()
maximum = sales[0]

for amount in sales:
    if amount > maximum:
        maximum = amount

print("Maximum Sales:", maximum)
```

**Question 8**: Given

prices = [500, 1500, 2500, 800, 1200]
1. Use map() + lambda to add 18% GST to all prices.
2. Use filter() to select prices above 1000.
3. Explain where such transformations are used in real-world pipelines.

```python
# Product prices
prices = [500, 1500, 2500, 800, 1200]

# 1. Add 18% GST using map() and lambda
gst_prices = list(map(lambda x: x + (x * 0.18), prices))

print("Prices After Adding GST:")
print(gst_prices)

# 2. Filter prices above 1000
filtered_prices = list(filter(lambda x: x > 1000, prices))

print("\nPrices Above 1000:")
print(filtered_prices)
```

Explain where such transformations are used in real-world pipelines.

-> Such transformations are widely used in real-world data pipelines for data preprocessing and analysis.

- map() is used to transform data, such as applying taxes, scaling values, or converting units.
- filter() is used to remove unwanted data and select relevant records.
- These operations help clean and prepare data before machine learning and analytics tasks.
- They improve efficiency when handling large datasets in business and Data Science applications.

**Question 9**:
Write a generator that yields squares of numbers from 1 to 1,000,000.

Explain difference between:

- List comprehension

- Generator expression

Why are generators preferred in big data scenarios?

```python
# Generator function for squares

def square_generator():
    for i in range(1, 1000001):
        yield i * i

# Create generator
squares = square_generator()

# Print first 5 square values
for _ in range(5):
    print(next(squares))
```

List Comprehension
- Creates and stores all values in memory at once.
- Uses square brackets [ ].
- Faster for small datasets.

Generator Expression
- Generates values one at a time when needed.
- Uses parentheses ( ).
- Saves memory and is efficient for large datasets.

Why are generators preferred in big data scenarios?

-> Generators are preferred in big data applications because they use less memory by producing values only when required.
- They handle large datasets efficiently.
- They are useful in streaming and machine learning workflows where continuous data processing is needed.
- They improve performance in data processing pipelines.

**Question 10**: Design a class:
class Product:

With:

- Attributes: name, price, quantity
- Method: calculate_revenue()

Create 3 objects and print their revenue.

```python
# Define Product class
class Product:

    # Constructor
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    # Method to calculate revenue
    def calculate_revenue(self):
        return self.price * self.quantity

# Create objects
product1 = Product("Laptop", 50000, 5)
product2 = Product("Phone", 20000, 10)
product3 = Product("Tablet", 15000, 7)

# Print revenue
print(product1.name, "Revenue:", product1.calculate_revenue())
print(product2.name, "Revenue:", product2.calculate_revenue())
print(product3.name, "Revenue:", product3.calculate_revenue())
```
